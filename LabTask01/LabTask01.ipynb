### Part 7: Mapping and Transformation ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).



```python
df['Sex_code'] = df['Sex'].map({'male': 0, 'female': 1})
df.head()
```

```python
def age_group(age):
    if pd.isna(age):
        return 'Unknown'
    elif age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

df['Age_group'] = df['Age'].apply(age_group)
df.head()
```

1. **`df['Sex'].map({'male': 0, 'female': 1})`** replaces *string values* with numeric codes using a ***dictionary***.
2. **`df['Age'].apply(age_group)`** applies a *custom function* to each Age value, returning **Child**, **Adult**, **Senior**, or **Unknown**.

### Part 8:

```python
import time
import csv
import urllib.request

url = "https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv"
with urllib.request.urlopen(url) as response:
    lines = response.read().decode('utf-8').splitlines()

titanic_list = list(csv.DictReader(lines))
```

Task 1 - Select Name column

```python
# Part 2, Task 1 - Select the Name column
start = time.perf_counter()
result = df['Name']
pandas_time = time.perf_counter() - start

start = time.perf_counter()
result = [row['Name'] for row in titanic_list]
python_time = time.perf_counter() - start

print(f"Pandas: {pandas_time:.6f}s | Python: {python_time:.6f}s")
```

Task 2 - Filter Age > 60

```python
# Part 3, Task 1 — Filter rows where Age > 60
start = time.perf_counter()
result = df[df['Age'] > 60]
pandas_time = time.perf_counter() - start

start = time.perf_counter()
result = [row for row in titanic_list if row['Age'] != '' and float(row['Age']) > 60]
python_time = time.perf_counter() - start

print(f"Pandas: {pandas_time:.6f}s | Python: {python_time:.6f}s")
```

Task 3 - Convert to Euros

```python
# Part 5, Task 1 — Convert Fare to Euros
start = time.perf_counter()
result = df['Fare'] * 0.92
pandas_time = time.perf_counter() - start

start = time.perf_counter()
result = [float(row['Fare']) * 0.92 for row in titanic_list]
python_time = time.perf_counter() - start

print(f"Pandas: {pandas_time:.6f}s | Python: {python_time:.6f}s")
```
